```python
import json
try:
    rfile=open('data/usagov_bitly.txt','r',encoding="utf8")
    lines=rfile.readlines()     #lines 리스트 줄단위로 읽기
    #print(type(lines),lines)
    #json디코딩: json.loads()
    #lines-> [json문자열들..]->파이썬 obj(dict) 객체
    rows=[]
    for row in lines[:10]:
        rows.append(json.loads(row))
    # rows=[json.loads(row) for row in lines]
    print('rows : ',len(rows))
    print('rows[0]의 타입:',type(rows[0]))
    for row in rows:
        print(row['a'])
except Exception as e:
    print("Error발생 : ",e)
finally:
    rfile.close()


## dict->DataFrame변환
import pandas as pd
record_df=pd.DataFrame(rows)
print(record_df.info())
record_df
```

```python
# pandas : DataFrame의 구조 (Table과 같은 구조)
# Series 들이 결합되어 있는 형태, Series 컬럼을 기반으로 데이터가 저장되어 있는 형태
# Pandas에서 참조시에는 Series의 이름을 사용하여 참조...

# Series의 특징은 List, Dictionary와 같은 성격을 가지고 있다.

# Series -> 판다스 기본 데이터 구조 (비슷, 리스트와 비슷함..)
#           리스트와 차이점은 인덱싱 방식에 따라 리스트와 다른 인덱싱 가능
ls=[10,20,30,40,50]
print(ls)

s=pd.Series(ls,name='lst')

#for으로 확인 list와 같은 동작
for i in s:
    print(i)

for r in range(len(s)):
    print(s[r])
print(s[0:2])       #슬라이싱 가능
s.info()

#리스트와 다른점 -> 인덱싱 처리
s1=pd.Series(ls, index=[1,2,3,4,5],name='index수정')
s1.info()
s1
```

```python
'''
<문제>
다음의 <보기> 자료를 딕셔너리를 기초자료로 하여 s1_jumsu시리즈를 만드세요
s1_jumsu 시리즈는 데이터(요소)가 평균이고 인덱스가 이름으로 구성되도록 하세요
Series를 만들면 이것을 <실행결과>와 같이 출력하세요.

<보기>
jumsu={
    '홍길동' : [50,60,50],
    '이순신' : [60,60,60],
    '강감찬' : [100,90,90],
    '유관순' : [80,80,90],
    '강환석' : [95,95,95]
}

<실행결과>
이름
점수평균
'''
from statistics import mean
jumsu={
    '홍길동' : [50,60,50],
    '이순신' : [60,60,60],
    '강감찬' : [100,90,90],
    '유관순' : [80,80,90],
    '강환석' : [95,95,95]
}
jumsu_s=pd.Series(jumsu)
jumsu_s
avg=[]
for i in range(len(jumsu_s)):
    avg.append(mean(jumsu_s[i]))
name=list(jumsu.keys())
print(type(name))
jumsu_s2={name[i]:avg[i] for i in range(len(jumsu))}
for key in jumsu_s2:
    print(key,end='\t')
print()
for key in jumsu_s2:
    print(jumsu_s2[key],end='\t')
```

```python
jumsu={
    '홍길동' : [50,60,50],
    '이순신' : [60,60,60],
    '강감찬' : [100,90,90],
    '유관순' : [80,80,90],
    '강환석' : [95,95,95]
}

# 4개의 Series로 만들어 주세요
# 이름(name_s), 각 리스트 0번은 국어(kor_s), 리스트 1번은 영어(eng_s),
# 리스트 2번은 수학(math_s)

name=[key for key in jumsu]
name_s=pd.Series(name)
kor=[jumsu[key][0] for key in jumsu]
eng=[jumsu[key][1] for key in jumsu]
math=[jumsu[key][2] for key in jumsu]
kor_s=pd.Series(kor)
eng_s=pd.Series(eng)
math_s=pd.Series(math)

#DataFrame만들기
score=pd.DataFrame({'이름':name_s,'국어':kor_s,'영어':eng_s,'수학':math_s})
score

#국영수 점수 합계 시리즈 생성
total=kor_s+eng_s+math_s
print(total)
tot=pd.DataFrame({'이름':name_s,'합계':total})

#평균값 구하기
avg=round(total/3,2)
avg_s=pd.DataFrame({'이름':name_s,'평균':avg})

score['합계']=total
score['평균']=avg
score
```
