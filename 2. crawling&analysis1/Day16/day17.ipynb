```python
#문제1
# 위의 temps1,temps2를 이용하여 뺄셈 외에 다른 연산 가능한지 실습
# 더하기,곱하기,나누기
sum=temps1+temps2
sum
mul=temps1*temps2
mul
div=temps1/temps2
round(div,2)
# 문제2
#temps3을 생성, temps1과 temps2에 대응하는 시리즈 생성
#온도 : 75,77,67,82,77,85,89, 인덱스 dates
temps3=pd.Series([75,77,67,82,77,85,89],index=dates)
temps3
#이렇게 생성된 temps1,temps2,temps3을 temps_df로 데이타 프레임으로 생성
#temps1-Missoula,temps2-Philadelphia,temps3-Seoul

temps_df=pd.DataFrame({'Missoula':temps1,'Philadelphia':temps2,'Seoul':temps3})
temps_df

# 문제3. 각 도의 평균온도 구하세요

missoula=round(temps_df.Missoula.mean(),2)
phil=round(temps_df.Philadelphia.mean(),2)
seoul=round(temps_df.Seoul.mean(),2)

print('missoula 평균 : ',missoula)
print('philadelphia 평균 : ', phil)
print('seoul 평균 : ', seoul)
```

```python
# 사용자 지정 인덱스로 찾기
# 1. 단일 : [데이터프레임 or Series].loc['label명',['컬럼명']]
#DataFrame
temps_df.loc['20220402']
temps_df.loc['20220401',['Missoula','Seoul']]
#Series
temps_df.Seoul.loc['20220401']

# 2. 여러개일 경우(특정 인덱스만 출력)
#DataFrame
temps_df.loc[['20220401','20220405'],['Philadelphia','Seoul']]
#Series
temps_df.Seoul.loc[['20220401','20220405']]

# 3. 여러개일 경우 (연속된 값)

#DataFrame
temps_df.loc['20220401':'20220405',['Philadelphia','Seoul']]
#Series
temps_df.Seoul.loc['20220401':'20220405']

# 4. Boolean 값을 이용한 출력 (이때에 인덱스 값과 같은 개수의 리스트 생성)
ls=[True,False,False,True,False,False,True]
#DataFrame
temps_df.loc[ls,['Philadelphia','Seoul']]
#Series
temps_df.Philadelphia.loc[ls]
```

```python
# 기본 인덱스 이용하여 접근
# 접근법
# 1. 단일 :[데이터프레임 or Series].iloc[index값(int)] -좌표값 찍는 방식
#DataFrame
temps_df.iloc[3,1]
#Series
temps_df.Philadelphia.iloc[3]

# 2. 여러개일 경우(특정 인덱스만 출력)
#DataFrame
temps_df.iloc[[1,3,5],[1,2]]    #[[값인덱스],[컬럼 인덱스]]
#Servies
temps_df.Philadelphia.iloc[[1,3,5]]

# 3. 여러개일 경우 (연속된 값인 경우)
#DataFrame
temps_df.iloc[1:4,[0,2]]
#Series
temps_df.Seoul.iloc[3:]

# 4. Boolean 이용한 출력
#DataFrame
temps_df.iloc[ls,[0,1,2]]
#Series
temps_df.Seoul.iloc[ls]
```
