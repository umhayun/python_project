```python
##bugs 크롤링
from selenium import webdriver
driver=webdriver.Chrome('c:\webdriver\chromedriver.exe')
url='https://music.bugs.co.kr/chart/track/realtime/total?wl_ref=M_contents_03_01'

driver.get(url)
```

```python
from bs4 import BeautifulSoup
html=driver.page_source
soup=BeautifulSoup(html,'html.parser')
```

```python
songs=soup.select('table.byChart>tbody>tr')
print(len(songs))
for song in songs:
    title=song.select('.title> a')[0].text
    artist=song.select('.artist >a')[0].text
    print(title,'\t\t\t', artist)
```

```python
song_data=[]
rank=1
for song in songs:
    title=song.select('.title>a')[0].text
    artist=song.select('.artist>a')[0].text
    song_data.append(['Bugs',rank,title,artist])
    rank+=1
song_data
```

```python
import pandas as pd
df_col=['서비스','순위','타이틀','가수']
pd_df=pd.DataFrame(song_data,columns=df_col)
pd_df
```

```python
pd_df.to_excel('data/bugs.xlsx',index=False)
```

```python
#genie 차트 알아오기...
url="https://www.genie.co.kr/chart/top200?ditc=D&ymd=20220518&hh=11&rtm=Y&pg=1"
driver=webdriver.Chrome('c:\webdriver\chromedriver.exe')
driver.get(url)
```

```python
from bs4 import BeautifulSoup

html=driver.page_source
soup=BeautifulSoup(html,'html.parser')
```

```python
songs=soup.select('.list-wrap>tbody>tr.list')
len(songs)
for song in songs:
    title=song.select('td.info>a.title')[0].text.strip()
    title=title.replace("19금","").strip()
    artist=song.select('td.info>a.artist')[0].text.strip()
    print(title,artist)
    
```

```python
song_data=[]
rank=1
for song in songs:
    title=song.select('td.info>a.title')[0].text.lstrip()
    title=title.replace("19금","").strip()
    artist=song.select('td.info>a.artist')[0].text.lstrip()
    song_data.append(['Genie',rank,title,artist])
    rank+=1
song_data
```

```python
import pandas as pd
col=['서비스','순위','타이틀','가수']
pd_df=pd.DataFrame(song_data,columns=col)
pd_df
```

```python
pd_df.to_excel('data/genie.xlsx',index=False)
```

```python
##Excel_Merge
excel_names=['data/bugs.xlsx','data/genie.xlsx','data/melon.xlsx']

appended_data=pd.DataFrame()
for name in excel_names:
    pd_df=pd.read_excel(name)
    appended_data=appended_data.append(pd_df)
```
